```python
def download_data(name):
    pd_matr = storage.show_chain_md_matrices().to_pandas()
    bzr = pd_matr.loc[pd_matr['sel_name'] == name + '.sdf']
    bzr_ind = bzr.id.to_list()
    for i in bzr_ind:
        bzr_load = ChainDescriptorsCB.load(storage, i)
        matrix = bzr_load.descr.md_mat_[0]
        file = 'data_varvara/' + name + '/'
        file = file + name + '_matrix_alphabet_NN'
        if bzr.loc[bzr['id'] == i, ['d_marker']].iloc[0, 0] == True:
            file += 'd'
        else:
            file += '*'
        if bzr.loc[bzr['id'] == i, ['b_marker']].iloc[0, 0] == True:
            file += 'b'
        else:
            file += '*'
        if bzr.loc[bzr['id'] == i, ['r_marker']].iloc[0, 0] == True:
            file += 'r'
        else:
            file += '*'
        file += '_UpTo' + str(bzr.loc[bzr['id'] == i, ['length']].iloc[0, 0]) + 'chains.csv'
        matrix.to_csv(file, header=True, index=False, sep=',')
        
    data = storage.show_datasets().to_pandas()
    bzr_ind_name_y = data.loc[data['name'] == name + '.sdf', ['id', 'prop', 'name']]
    bzr_y_load = SDFDatasetCB.load(storage, bzr_ind_name_y.id)
    bzr_y = bzr_y_load.view.table_[bzr_ind_name_y.prop]
    f = 'data_varvara/' + name + '/' + name + '_y.csv'
    bzr_y.to_csv(f, header=True, index=False, sep=',')
```
